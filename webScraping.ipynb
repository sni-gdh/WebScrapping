```python
import os
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
```

```python
BASE_URL = "https://huggingface.co"
Source_URL = "https://huggingface.co/docs/transformers/index"
```

```python
headers = {
    "User-Agent": "Mozilla/5.0"
}
```

```python
MAIN_FOLDER = "Vision Models"
os.makedirs(MAIN_FOLDER, exist_ok=True)
```

```python
response = requests.get(Source_URL, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")
```

```python
print(soup)
```

```python
model_link = None

for link in soup.find_all("a", href=True):
    if "models" in link.text.lower():
        model_link = urljoin(BASE_URL, link["href"])
        break

if not model_link:
    print("model section not found")

print("model page:", model_link)
```

```python
response = requests.get(model_link, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")
```

```python
main_section = soup.find("main")
```

```python
vision_models = {}

for link in main_section.find_all("article"):
    model_name = link.find("h4").text.lower()
    vision_models[model_name] = urljoin(BASE_URL, link.find("a")["href"])

if not vision_models:
    print("Vision models page not found")

print("Number of models:",vision_models)
```

```python
for model, url in vision_models.items():
    print("Scraping:", model)
    
    model_folder = os.path.join(MAIN_FOLDER, model)
    os.makedirs(model_folder, exist_ok=True)
    
    try:
        r = requests.get(url, headers=headers)
        page = BeautifulSoup(r.text, "html.parser")
    
        main_content = page.find("main")
        for val in main_content.find_all("section"):
            main_content = val
            break
    
        if main_content:
            text = main_content.get_text(separator="")
    
            file_path = os.path.join(model_folder, "info.txt")
    
            with open(file_path, "w", encoding="utf-8") as f:
                f.write(text)
    
    except Exception as e:
        print("Error scraping", model, e)

print("Scraping finished")
```
